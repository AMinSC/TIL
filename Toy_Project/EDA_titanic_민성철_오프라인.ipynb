# 탑승 위치 별 생존률

```python
# 위치별 탑승률

df['embarked'].value_counts()
```

```python
# 위치별 평균치

emb_df = df.groupby('embarked').mean()
emb_df
```

```python
# 위치별 생존률

df[['embarked', 'survived']].value_counts()
```

```python
# 위치별 생존률, 평균치
how = {
    "survived" : np.mean
}
t = df.groupby('embarked').agg( how )
t
```

```python
#위치별 평균치 생존률 시각화
t.plot.bar()
```

```python
# 시각화 데이터를 참고하여 추가적인 데이터 확인 (pclass, sex,..)
plt.figure(figsize=(15, 10))
sns.barplot(x='embarked', y='survived', data=df, hue='sex')
```

탑승 위치별 확인되는 특성(age, pclass, fare)으로 보아 생존률과 관계가 있는것이 확인된다.
예를 들어 C 인 Cherbourg의 경우, 평균치로 age와 fare가 가장높으며 생존률 또한 타 지역보다 상대적으로 높은게 확인된다

# Pclass, sex별  생존률 비교

```python

df[['pclass', 'age', 'embarked', 'sex', 'survived']]
```

```python
# pclass별 여성 사망률
fe_sur = df[(df['sex']=='female') & (df['survived']==0)][['pclass']].value_counts().to_frame()
fe_sur
```

```python
# pclass별 남성 생존률
male_sur = df[(df['sex']=='male') & (df['survived']==1)][['pclass']].value_counts().to_frame()
male_sur
```

```python
# 위 두 데이터를 merge로 합친뒤 column명 변경

sur_df = pd.merge(fe_sur, male_sur, left_on='pclass', right_on='pclass')
sur_df.columns
sur_df.rename(columns={'0_x':'female_die', 
                       '0_y':'male_liv'}, inplace=True)
sur_df
```

```python
# 1, 위 데이터 시각화 / 2, pclass별 sex 생존률 / 3, 전체 sex별 생존률

fig = plt.figure(figsize=(16 ,10))
graph1 = fig.add_subplot(1,3,1)
graph2 = fig.add_subplot(1,3,2)
graph3 = fig.add_subplot(1,3,3)

sur_df.plot.bar(ax=graph1)
sns.barplot('pclass', 'survived', hue='sex', data=df, ax=graph2)
sns.barplot('sex', 'survived', data=df, ax=graph3)
```

전반적으로 female의 생존률이 높게 확인되며, pclass가 1에 가까울 수록 생존률이 높아지는것도 확인된다.
